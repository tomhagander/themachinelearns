```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# read in the csv file and store the values in lists
housing_data = []
with open('Houses.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        housing_data.append([float(row['Monthly energy cost in winter ($)']), float(row['Monthly energy cost in summer ($)'])])

# convert the lists to numpy arrays
housing_data = np.array(housing_data)

# run k-means clustering
# set the number of clusters to 3
kmeans = KMeans(n_clusters=3).fit(housing_data)

# plot the results
plt.scatter(housing_data[:,0], housing_data[:,1], c=kmeans.labels_, cmap='rainbow')
```
